Let's see how correlated our numeric data is but we will do so using a heatmap to have a clear look

```python
# Due to recent changes in panda we will have to include only the numberic data for the correlation matrix
df_numeric = df.select_dtypes(include='number')
df_numeric.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.show
```

We can see that all the population columns are highly correlated with each other with makes sense since the population tends to increase over time especially if wars were avoided, I also thought of pandemics but it doesn't seem to have that large of an effect on the population considering the fact tha Covid happened in late 2020 and the correlation remained pretty high and the same thing can be said about the percentage too. The Area(km2) also have a positive correlation with the population but it's not that high, on the other hand the density have a weak to negative correlation with population wich means that high population doesn't necessarily mean a highly dense country especially if said country has such a large area, as for the growth rate it's a percentage so it's normal for it to not have a strong correlation with the population, normally the growth rate in large countries tends to be more stabe and steady.

Now let's look at the population per continent

```python
# Due to the changes in pandas we will have to filter the columns we want to include in our groupby as well as the column we will use to groupby
# Let's just drop the irrelevant columns for this step
df_filtered = df.drop(['Rank', 'CCA3', 'Country', 'Capital'], axis=1).set_index('Continent')
df_filtered
```

```python
# We group by Continent
df_filtered.groupby('Continent').sum().sort_values(by='2022 Population', ascending=False)
```

Unsurprisingly, Asia ranks first in nearly all features. While the growth rate is a percentage, meaning we cannot draw meaningful conclusions based on the sum of all rates per continent, other indicators clearly demonstrate Asia's dominance.
Since 1970, Asia has consistently had the highest population, by a substantial margin, which is also reflected in its contribution to the world population percentage — making up nearly 60%. Additionally, Asia has the largest area and the highest population density, further highlighting its significant global presence.

Now let's visualize the population columns

```python
# We filter again to include only the population columns and then we transpose it for better readability
df_population = df_filtered.drop(['Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage'], axis=1)
df_population.groupby('Continent').sum().transpose().sort_values(by='Africa', ascending=True)
# I sorted the values so that we I plot the data it will the chronological order starting from 1970
```

```python
df_population.groupby('Continent').sum().transpose().sort_values(by='Africa', ascending=True).plot(figsize=(15,7))
plt.title("Population Growth by Continent (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.grid(True)
plt.legend(title="Continent", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

This plot helps us visualize the difference there is between Asia and all the other continents in terms of population and further confirms what was deduced earlier from the data, as we can see Asia's population was pretty high and a lot larger than the other continents since 1970, and this gap kept on increasing over the years.

Next we will visualize the population but per country using a boxplot

```python
df_country = df.drop(['Rank', 'CCA3', 'Continent', 'Capital', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage'], axis=1)
df_country.groupby('Country').sum()
```

```python
df_country.groupby('Country').sum().boxplot(figsize=(15,7))
plt.xlabel("Year")
plt.title("Population distribution per year (1970-2022)")
plt.ylabel("Total Population per Country")
plt.show()
```

We can see from the plot that there are two outliers no matter the year, we can confidently say that these two countries are China and India.
